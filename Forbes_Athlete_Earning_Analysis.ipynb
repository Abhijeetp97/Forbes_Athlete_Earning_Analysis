Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020)

```python
import pandas as pd
df_forbes = pd.read_csv('/content/Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020).csv')

df_forbes
```

```python
print(df_forbes.info())
print(df_forbes.describe())
```

```python
print(df_forbes.isnull().sum())
```

```python
#Top Earners by Year
top_earners_by_year = df_forbes.loc[df_forbes.groupby('Year')['earnings ($ million)'].idxmax()]

print(top_earners_by_year)
```

```python

most_common_sports = df_forbes['Sport'].value_counts()#Earnings Trend Over Time
earnings_trend = df_forbes.groupby('Year')['earnings ($ million)'].sum()

print(earnings_trend)
```

```python
#Nationality Analysis
nationality_counts = df_forbes['Nationality'].value_counts()

print(nationality_counts)
```

```python
#Rank Changes
df_forbes['Rank Change'] = pd.to_numeric(df_forbes['Previous Year Rank'], errors='coerce').fillna(0).astype(int) - pd.to_numeric(df_forbes['Current Rank'], errors='coerce').fillna(0).astype(int)
# Use pd.to_numeric to handle non-numeric values, setting them to NaN with errors='coerce'
# Fill NaN values with 0 before converting to integers

print(df_forbes[['Name', 'Year', 'Rank Change']])
```
